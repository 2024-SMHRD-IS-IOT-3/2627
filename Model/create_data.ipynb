```python
import pandas as pd
import numpy as np
import os
```

# 데이터 불러오기

```python
path = './air/'
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.csv')] ## 파일명 끝이 .csv인 경우

print(file_list_py)

## csv 파일들을 DataFrame으로 불러와서 concat

df_air = pd.DataFrame()
for i in reversed(file_list_py):
    data = pd.read_csv(path + i, encoding='cp949')
    df_air = pd.concat([df_air,data])

df_air = df_air.reset_index(drop = True)

# df_sorted_asc = df.sort_values(by='date')
df_air['일시'] = pd.to_datetime(df_air['일시'])
df_air['일시'] = df_air.sort_values(by='일시')['일시'].reset_index(drop=True)
# df_air.set_index('일시', inplace=True)
df_air
```

```python
df_air = df_air[['일시','1시간평균 미세먼지농도(㎍/㎥)']]
df_air
```

```python
df_solar_train_data = pd.read_csv('solar_traindata.csv', encoding='cp949')

df_solar_train_data['태양고도각'] = round(df_solar_train_data['태양고도각'],1)
df_solar_train_data['태양방위각'] = round(df_solar_train_data['태양방위각'],1)
df_solar_train_data['발전량(Target)'] = round(df_solar_train_data['발전량(Target)'],1)


# df_solar_train_data 결측치 확인
df_solar_train_data
```

# 날짜 기준으로 데이터 통합하기

- 모든 데이터는 광주지역을 기준으로 작성되어있음


```python
# 태양광 발전량 17.01 - 23.11
# 시간 - 08:00 - 18:00
# 태양광 발전량 csv의 경우에는 모든 파일이 다 08 :00 - 18:00 까지만 표기
df_solar_train_data.columns
df_solar = df_solar_train_data[['일시', '태양방위각', '태양고도각', '기온', '강수량', '풍속', '풍향',
       '습도', '증기압', '이슬점온도', '현지기압', '일조', '일사']]

solar_power = df_solar_train_data[['발전량(Target)']]
df_solar['일시'] = pd.to_datetime(df_solar['일시'])
df_solar
```

```python
train = pd.merge(df_solar, df_air, on='일시', how='outer')
train = pd.concat([train,solar_power], axis=1)

train.rename(columns={'1시간평균 미세먼지농도(㎍/㎥)':'미세먼지'}, inplace=True)

# train 데이터 결측치 확인
train.isnull().sum()


# 결측치가 많은 이유는 미세먼지의 경우는 24시간 시간별 데이터가 전부 존재하지만
# 태양광의 경우는 08시부터 18시의 데이터만 존재하기 때문에 따라서 우리 모델의 경우도 08시부터 18시에 태양광 발전량만 예측하는 모델로 변경하여 진행

# 08시부터 18시의 데이터만 남기고 모두 정리하기
train = train.dropna()
train.info()
```
